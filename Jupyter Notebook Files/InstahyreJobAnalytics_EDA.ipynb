```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/Users/Horse/Downloads/job_data - Copy.csv")
```

```python
df
```

```python
#Creating Tables

table1= pd.DataFrame()
table2= pd.DataFrame()
table3= pd.DataFrame()
```

```python
#Looking for Null values

df.isna().sum()
df=df.dropna()
df.isna().sum()
```

```python
#Creating Jobs table

table1['job_id']= df.job_id
table1['company_id']=0
table1['location']=df.job_locations
table1['designation']=df.job_title
table1['details_id']=0
```

```python
table1
```

```python
#Creating Companies table

table2['company_id']=0
table2['name']=df.company_name
table2['estab_year']=df.founded_in
table2['employees_count']=df.employees_count
table2
```

```python
#Generating primary key for Companies table

x=table2.name.unique()
cid=[]
for i in range(len(x)):
    cid.append(i)
x=list(x)
for i in range(734):
    a=table2.name.iloc[i]
    table2.company_id.iloc[i]=cid[x.index(a)]
table2
```

```python
#Creating Job details table

table3['details_id']=0
table3['skills']=df.skills
table3['hr_name']=df.hr_name
```

```python
#Generating primary key for Job details table

table3.skills.value_counts()
y=list(table3.skills.unique())
did=[]
for i in range(len(y)):
    did.append(i)
for i in range(734):
    a=table3.skills.iloc[i]
    table3.details_id.iloc[i]=did[y.index(a)]
table3
```

```python
#Linking tables

table1['company_id']=table2.company_id
table1
table1.details_id=table3.details_id
table2=table2.drop_duplicates()
```

```python
#Datatype conversions

table1.details_id=table1.details_id.astype(int)
table3.details_id=table3.details_id.astype(int)
table2.estab_year=table2.estab_year.astype(int)
```

```python
#Converting tables to CVS files

table1.to_csv('table1x.csv', index=False)
table2.to_csv('table2x.csv', index=False)
table3.to_csv('table3x.csv', index=False)
```
